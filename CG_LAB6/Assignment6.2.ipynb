```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create dataset
data = {
    "Subject": ["Math", "Science", "English", "History", "Computer"],
    "Score": [85, 90, 78, 88, 92]
}
df = pd.DataFrame(data)

# Plot bar chart
plt.figure(figsize=(7, 5))
sns.barplot(x="Subject", y="Score", data=df, palette="viridis")

# Add annotations
for index, value in enumerate(df["Score"]):
    plt.text(index, value + 1, str(value), ha='center', fontsize=12)

# Add grid, title, and labels
plt.title("Student Scores in Different Subjects")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.grid(axis='y')
plt.show()
```
