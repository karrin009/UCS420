```python
import numpy as np
import matplotlib.pyplot as plt

# Get user input for M
M = float(input("Enter a value for M: "))

# Generate x values
x = np.linspace(-10, 10, 100)

# Compute y values
y1 = M * x**2
y2 = M * np.sin(x)

# Plot both functions
plt.figure(figsize=(8, 5))
plt.plot(x, y1, label=r'$y = Mx^2$', color='blue', linestyle='-')
plt.plot(x, y2, label=r'$y = M\sin(x)$', color='red', linestyle='--')

# Add grid, title, and legend
plt.grid(True)
plt.title("Mathematical Functions Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```
