```python
import pandas as pd
df=pd.read_csv('company_sales_data.csv')
print(df.head())
```

```python
#cleaning the data
# Strip any leading/trailing whitespace from column names
df.columns = df.columns.str.strip()

# Print column names to verify
print("Column Names:", df.columns.tolist())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Line plot for total profit
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='month_number', y='total_profit', marker='o', color='blue')
plt.title("Total Profit of All Months")
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.grid(True)
plt.show()
```

```python
# Identify product columns (assuming they are all columns except 'Month' and 'Total_Profit')
product_columns = df.columns.difference(['Month', 'Total_Profit'])

# Melt the dataframe to long format for seaborn
df_melted = df.melt(id_vars=['month_number'], value_vars=product_columns, var_name='Product', value_name='Sales')

# Multi-line plot for all product sales
plt.figure(figsize=(12, 8))
sns.lineplot(data=df_melted, x='month_number', y='Sales', hue='Product', marker='o')
plt.title("Monthly Sales Data for Each Product")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend(title='Product', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Set 'Month' as the index
df.set_index('month_number', inplace=True)

# Bar chart for all attributes
df.plot(kind='bar', figsize=(14, 8), colormap='viridis')
plt.title("Sales Data for All Features")
plt.xlabel("month_number")
plt.ylabel("Values")
plt.grid(axis='y')
plt.legend(title='Attributes', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
