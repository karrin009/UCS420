```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed using roll number (replace with your actual roll number)
roll_number = 12345  # Replace this with your actual roll number
np.random.seed(roll_number)

# Generate dataset
data = np.random.randn(50)

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Line plot of cumulative sum
axes[0, 0].plot(np.cumsum(data), color='blue', label="Cumulative Sum")
axes[0, 0].set_title("Cumulative Sum Plot")
axes[0, 0].set_xlabel("Index")
axes[0, 0].set_ylabel("Cumulative Sum")
axes[0, 0].legend()
axes[0, 0].grid()

# Scatter plot with random noise
axes[0, 1].scatter(range(50), data, color='red', alpha=0.6, label="Random Noise")
axes[0, 1].set_title("Scatter Plot with Random Noise")
axes[0, 1].set_xlabel("Index")
axes[0, 1].set_ylabel("Value")
axes[0, 1].legend()
axes[0, 1].grid()

# Hide empty subplots
axes[1, 0].axis("off")
axes[1, 1].axis("off")

# Show plot
plt.tight_layout()
plt.show()
```
