Part I: Randomized Sales Data Genera on (NumPy)

```python
import numpy as np
import pandas as pd

# Step 1: Initialize Random Seed (Use your roll number)
roll_number = 102497007  
np.random.seed(roll_number)

# Step 2: Generate Sales Data (12 months, 4 categories)
sales_data = np.random.randint(1000, 5001, size=(12, 4))

# Step 3: Convert to DataFrame
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
categories = ["Electronics", "Clothing", "Home & Kitchen", "Sports"]

df = pd.DataFrame(sales_data, columns=categories, index=months)

# Display the DataFrame
print(df)
```

PART II-> Data Manipula on and Analysis (Pandas)

```python
# Step 1: Display first 5 rows and summary statistics
print(df.head())
print(df.describe())

# Step 2: Calculate total sales per category and per month
df["Total Sales"] = df.sum(axis=1)  # Sum across columns (monthly total)
category_totals = df.sum()  # Sum across rows (category total)
print("Total sales per category:\n", category_totals)
print("Total sales per month:\n", df["Total Sales"])

# Step 3: Calculate average sales growth per category
df["Growth Rate"] = df["Total Sales"].pct_change() * 100  # Percentage change
print("Growth Rate:\n", df["Growth Rate"])

# Step 5: Apply a randomized discount
if roll_number % 2 == 0:
    df["Electronics"] *= 0.90  # 10% discount
else:
    df["Clothing"] *= 0.85  # 15% discount
```

Part III: Visualiza ons (Matplotlib and Seaborn)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Line plot for sales trends
df.drop(columns=["Total Sales", "Growth Rate"]).plot(kind='line', marker='o', title="Monthly Sales Trends")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend(title="Category")
plt.grid()
plt.show()


# Box plot for sales distribution
plt.figure(figsize=(8,6))
sns.boxplot(data=df.drop(columns=["Total Sales", "Growth Rate"]))
plt.title("Sales Distribution by Category")
plt.show()
```
